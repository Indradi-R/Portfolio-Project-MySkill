### Machine Learning

```python
#data data yang sifatnya numerikal
new_data.select_dtypes(include = 'number')
```

```python
#membuat function yang bisa diterapkan di data testing
def select_columns(df: pd.DataFrame, features: list, target: list = 'rating'): #dikasih default parameter
  X = df[features]
  y = (df["rating"] >= 3).astype(int) #diubah jadi int

  return X, y #untuk membagi data kita
```

```python
from sklearn.model_selection import train_test_split
```

```python
#mengambil features untuk belajar
features = ["budget", "popularity", "runtime", "vote_average"]
X, y = select_columns(new_data, features = features)
X
```

```python
y
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
minmax.fit(X_train)
```

```python
#mengaplikasikan xtrainnya
X_train_scaled = minmax.transform(X_train)
X_test_scaled = minmax.transform(x_test)
```

```python
def scaling(df, scaler):
  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns = scaler.get_feature_names_out())

  return scaled_df
```

```python
scaling(X_train, minmax)
```

```python
scaling(x_test, minmax)
```
